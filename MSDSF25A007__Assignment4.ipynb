```python
# Custom Exception for insufficient funds
class InsufficientFundsError(Exception):
    """Raised when withdrawal amount exceeds available balance"""
    def __init__(self, balance, amount):
        self.balance = balance
        self.amount = amount
        super().__init__(f"Insufficient funds: Trying to withdraw ${amount} but balance is ${balance}")

def perform_transaction(balance, amount, transaction_type):
    """
    Perform a bank transaction (deposit or withdraw)

    Args:
        balance (float): Current account balance
        amount (float): Transaction amount
        transaction_type (str): 'deposit' or 'withdraw'

    Returns:
        float: New balance after transaction

    Raises:
        InsufficientFundsError: If withdrawal exceeds balance
        ValueError: If invalid transaction type
    """
    try:
        # Validate transaction type
        if transaction_type not in ['deposit', 'withdraw']:
            raise ValueError(f"Invalid transaction type: '{transaction_type}'. Must be 'deposit' or 'withdraw'")

        # Validate amount
        if amount <= 0:
            raise ValueError(f"Amount must be positive: ${amount}")

        # Perform transaction
        if transaction_type == 'deposit':
            new_balance = balance + amount
            print(f"Deposited: ${amount}")

        elif transaction_type == 'withdraw':
            if amount > balance:
                raise InsufficientFundsError(balance, amount)
            new_balance = balance - amount
            print(f"Withdrawn: ${amount}")

    except (InsufficientFundsError, ValueError) as e:
        # Re-raise the exception to be handled by caller
        raise e
    else:
        # This runs only if no exception was raised
        print(f"Transaction successful! New balance: ${new_balance:.2f}")
        return new_balance
    finally:
        # This always runs, regardless of exception
        print("Transaction attempt completed\n")

def process_transactions(transactions_list, initial_balance=0):
    """
    Process a list of transactions sequentially

    Args:
        transactions_list: List of tuples [(type, amount), ...]
        initial_balance: Starting balance

    Returns:
        float: Final balance after all transactions
    """
    balance = initial_balance
    print(f"Starting balance: ${balance:.2f}")
    print("=" * 40)

    for i, (transaction_type, amount) in enumerate(transactions_list, 1):
        print(f"Transaction {i}: {transaction_type} ${amount}")

        try:
            balance = perform_transaction(balance, amount, transaction_type)

        except InsufficientFundsError as e:
            print(f"‚ùå Transaction failed: {e}")
            print("Skipping this transaction, continuing with next...\n")

        except ValueError as e:
            print(f"‚ùå Invalid transaction: {e}")
            print("Skipping this transaction, continuing with next...\n")

    return balance

# Demonstration with comprehensive examples
if __name__ == "__main__":
    print("üè¶ BANK TRANSACTION SYSTEM")
    print("=" * 50)

    # Test Case 1: Normal operations
    print("TEST CASE 1: Normal Transactions")
    transactions1 = [
        ('deposit', 1000),
        ('withdraw', 200),
        ('deposit', 500),
        ('withdraw', 300)
    ]
    final_balance1 = process_transactions(transactions1, 100)
    print(f"‚úÖ Final balance: ${final_balance1:.2f}")
    print("=" * 50)

    # Test Case 2: With error handling
    print("\nTEST CASE 2: Transactions with Errors")
    transactions2 = [
        ('deposit', 500),
        ('withdraw', 200),
        ('withdraw', 800),  # This should fail - insufficient funds
        ('deposit', 100),
        ('transfer', 50),   # This should fail - invalid type
        ('withdraw', -100), # This should fail - negative amount
        ('deposit', 300)
    ]
    final_balance2 = process_transactions(transactions2)
    print(f"‚úÖ Final balance: ${final_balance2:.2f}")
    print("=" * 50)

    # Test Case 3: Individual transaction demonstration
    print("\nTEST CASE 3: Individual Transaction Demo")
    try:
        balance = 1000
        print(f"Initial balance: ${balance}")

        # Successful deposit
        balance = perform_transaction(balance, 500, 'deposit')

        # Successful withdrawal
        balance = perform_transaction(balance, 200, 'withdraw')

        # Failed withdrawal - insufficient funds
        balance = perform_transaction(balance, 2000, 'withdraw')

    except (InsufficientFundsError, ValueError) as e:
        print(f"‚ùå Critical error: {e}")
    finally:
        print(f"Account status check complete. Current balance: ${balance}")
```

```python
# Custom Exception for invalid marks
class InvalidMarksError(Exception):
    """Raised when marks are outside valid range (0-100)"""
    def __init__(self, subject, marks):
        self.subject = subject
        self.marks = marks
        super().__init__(f"Marks for {subject} are invalid (value = {marks})")

def calculate_grade(marks_dict):
    """
    Calculate average marks and grade from a dictionary of subjects and marks

    Args:
        marks_dict (dict): Dictionary with subject names as keys and marks as values

    Returns:
        tuple: (average_marks, grade)

    Raises:
        InvalidMarksError: If any marks are outside 0-100 range
    """
    valid_marks = {}
    invalid_subjects = []

    # Validate each subject's marks
    for subject, marks in marks_dict.items():
        try:
            if marks < 0 or marks > 100:
                raise InvalidMarksError(subject, marks)
            valid_marks[subject] = marks
            print(f"‚úÖ {subject}: {marks} (Valid)")

        except InvalidMarksError as e:
            print(f"‚ùå Error: {e}")
            invalid_subjects.append(subject)

    # Check if we have any valid marks to process
    if not valid_marks:
        raise InvalidMarksError("All subjects", "No valid marks found")

    # Calculate average and grade
    try:
        average_marks = sum(valid_marks.values()) / len(valid_marks)

        # Determine grade
        if average_marks >= 90:
            grade = 'A'
        elif average_marks >= 80:
            grade = 'B'
        elif average_marks >= 70:
            grade = 'C'
        elif average_marks >= 60:
            grade = 'D'
        else:
            grade = 'F'

        return average_marks, grade

    except Exception as e:
        print(f"Unexpected error during calculation: {e}")
        return 0, 'F'

def process_student_grades(student_data):
    """
    Process grades for multiple students with comprehensive error handling
    """
    print("üéì STUDENT GRADING SYSTEM")
    print("=" * 50)

    for student_name, marks_dict in student_data.items():
        print(f"\nProcessing: {student_name}")
        print("-" * 30)

        try:
            average, grade = calculate_grade(marks_dict)
            print(f"\nüìä RESULTS for {student_name}:")
            print(f"   Average Marks: {average:.2f}")
            print(f"   Grade: {grade}")
            print(f"   Valid Subjects Processed: {len(marks_dict) - len([m for m in marks_dict.values() if m < 0 or m > 100])}")

        except InvalidMarksError as e:
            print(f"üö´ Cannot calculate grade for {student_name}: {e}")
        except Exception as e:
            print(f"üí• Unexpected error processing {student_name}: {e}")
        finally:
            print("Processing complete for this student")
            print("=" * 50)

# Alternative function that processes all valid subjects and ignores invalid ones
def calculate_grade_robust(marks_dict):
    """
    A more robust version that processes valid subjects and ignores invalid ones
    """
    valid_marks = {}

    print(f"\nProcessing {len(marks_dict)} subjects...")

    for subject, marks in marks_dict.items():
        try:
            if marks < 0 or marks > 100:
                raise InvalidMarksError(subject, marks)
            valid_marks[subject] = marks

        except InvalidMarksError as e:
            print(f"‚ö†Ô∏è  Skipping {subject}: {e}")
            continue  # Skip invalid subjects but continue processing

    if not valid_marks:
        return None, None  # No valid marks to process

    # Calculate average from valid marks only
    average_marks = sum(valid_marks.values()) / len(valid_marks)

    # Grade calculation
    if average_marks >= 90:
        grade = 'A'
    elif average_marks >= 80:
        grade = 'B'
    elif average_marks >= 70:
        grade = 'C'
    elif average_marks >= 60:
        grade = 'D'
    else:
        grade = 'F'

    return average_marks, grade, valid_marks

# Demonstration
if __name__ == "__main__":
    # Test data with various scenarios
    student_data = {
        "Ali Khan": {
            'Math': 85,
            'Science': 92,
            'English': 78,
            'History': 88
        },
        "Sara Ahmed": {
            'Math': 95,
            'Science': 120,  # Invalid - above 100
            'English': 82,
            'Physics': 105   # Invalid - above 100
        },
        "Hassan Raza": {
            'Math': -5,      # Invalid - negative
            'Science': 45,
            'English': 150   # Invalid - above 100
        },
        "Zara Malik": {
            'Math': 75,
            'Science': 81,
            'English': 79,
            'Biology': 86
        },
        "Bilal Shah": {
            'Math': 200,     # Invalid - above 100
            'Science': -10   # Invalid - negative
        }
    }

    # Process all students
    process_student_grades(student_data)

    # Demonstrate individual robust processing
    print("\n" + "=" * 60)
    print("INDIVIDUAL ROBUST PROCESSING DEMO")
    print("=" * 60)

    test_cases = [
        {'Math': 85, 'Science': 92, 'English': 78},
        {'Math': 95, 'Science': 120, 'English': 82},
        {'Math': -5, 'Science': 45, 'English': 150},
        {'Math': 75, 'Science': 81, 'Chemistry': 110}
    ]

    for i, marks_dict in enumerate(test_cases, 1):
        print(f"\nTest Case {i}: {marks_dict}")
        result = calculate_grade_robust(marks_dict)

        if result[0] is not None:
            avg, grade, valid_subjects = result
            print(f"‚úÖ Valid subjects: {list(valid_subjects.keys())}")
            print(f"‚úÖ Average: {avg:.2f}, Grade: {grade}")
        else:
            print("‚ùå No valid subjects to process")
```

```python
# Custom Exception for invalid data
class InvalidDataError(Exception):
    """Raised when data line cannot be parsed or is invalid"""
    def __init__(self, line, reason):
        self.line = line
        self.reason = reason
        super().__init__(f"Invalid data: '{line}' - {reason}")

def parse_data(data_lines):
    """
    Parse student data lines and convert to dictionary

    Args:
        data_lines (list): List of strings in format "name,score"

    Returns:
        dict: Dictionary mapping student names to scores
    """
    parsed_data = {}
    valid_count = 0
    invalid_count = 0

    print("üìù PARSING STUDENT DATA")
    print("=" * 50)

    for i, line in enumerate(data_lines, 1):
        try:
            # Basic validation - check if line is empty
            if not line.strip():
                raise InvalidDataError(line, "Empty line")

            # Split the line
            parts = line.strip().split(',')

            # Check if we have exactly 2 parts
            if len(parts) != 2:
                raise InvalidDataError(line, f"Expected 2 parts, got {len(parts)}")

            name, score_str = parts

            # Validate name
            name = name.strip()
            if not name:
                raise InvalidDataError(line, "Empty student name")

            # Validate and convert score
            try:
                score = int(score_str.strip())
            except ValueError:
                raise InvalidDataError(line, f"Score '{score_str}' is not a valid integer")

            # Check score range
            if score < 0 or score > 100:
                raise InvalidDataError(line, f"Score {score} out of range (0-100)")

            # Check for duplicate names
            if name in parsed_data:
                print(f"‚ö†Ô∏è  Warning: Duplicate entry for '{name}', keeping latest score")

            # All validations passed - add to dictionary
            parsed_data[name] = score
            valid_count += 1
            print(f"‚úÖ Line {i}: '{name}' -> {score}")

        except InvalidDataError as e:
            invalid_count += 1
            print(f"‚ùå Line {i}: {e}")

        except Exception as e:
            invalid_count += 1
            print(f"üí• Line {i}: Unexpected error - {e}")

    print("=" * 50)
    print(f"üìä PARSING SUMMARY:")
    print(f"   Valid entries: {valid_count}")
    print(f"   Invalid entries: {invalid_count}")
    print(f"   Total processed: {len(data_lines)}")
    print(f"   Final dictionary size: {len(parsed_data)}")

    return parsed_data

def get_top_student(parsed_data):
    """
    Find the student with the highest score

    Args:
        parsed_data (dict): Dictionary of student names to scores

    Returns:
        tuple: (student_name, score) or (None, None) if no data
    """
    try:
        if not parsed_data:
            print("üì≠ No valid student data available")
            return None, None

        # Find student with maximum score
        top_student = max(parsed_data.items(), key=lambda x: x[1])

        print(f"üèÜ TOP STUDENT: {top_student[0]} with score {top_student[1]}")
        return top_student

    except Exception as e:
        print(f"üí• Error finding top student: {e}")
        return None, None

def analyze_student_performance(parsed_data):
    """
    Additional analysis function to demonstrate data processing
    """
    if not parsed_data:
        print("No data to analyze")
        return

    print("\nüìà PERFORMANCE ANALYSIS")
    print("-" * 30)

    # Calculate statistics
    scores = list(parsed_data.values())
    average = sum(scores) / len(scores)
    max_score = max(scores)
    min_score = min(scores)

    print(f"Total Students: {len(parsed_data)}")
    print(f"Average Score: {average:.2f}")
    print(f"Highest Score: {max_score}")
    print(f"Lowest Score: {min_score}")

    # Grade distribution
    grades = {'A': 0, 'B': 0, 'C': 0, 'D': 0, 'F': 0}
    for score in scores:
        if score >= 90: grades['A'] += 1
        elif score >= 80: grades['B'] += 1
        elif score >= 70: grades['C'] += 1
        elif score >= 60: grades['D'] += 1
        else: grades['F'] += 1

    print("\nGrade Distribution:")
    for grade, count in grades.items():
        if count > 0:
            print(f"  {grade}: {count} student(s)")

# Demonstration with various test cases
if __name__ == "__main__":
    # Test data with various error scenarios
    test_data = [
        "Ali,85",
        "Sara,90",
        "Bilal,invalid",
        "Hassan,78",
        "Zara,102",
        "Ahmed,-5",
        "Fatima,95",
        "John  Doe,88",  # Missing comma
        " ,75",          # Empty name
        "Ayesha,85",
        "Ali,92",        # Duplicate name
        "Raza,seventy",  # Invalid score format
        "Bilal,88",      # Previously invalid, now valid
        "",
        "Extra,Comma,Here",  # Too many commas
        "Final,100"
    ]

    print("üéì STUDENT DATA PROCESSOR")
    print("=" * 60)

    # Parse the data
    student_scores = parse_data(test_data)

    # Get top student
    print("\n" + "=" * 40)
    top_student, top_score = get_top_student(student_scores)

    # Additional analysis
    if student_scores:
        analyze_student_performance(student_scores)

    # Display final results
    print("\n" + "=" * 40)
    print("üìã FINAL STUDENT SCORES:")
    if student_scores:
        for name, score in sorted(student_scores.items()):
            print(f"   {name}: {score}")
    else:
        print("   No valid student data available")

    # Test with empty data
    print("\n" + "=" * 60)
    print("TEST WITH EMPTY DATA:")
    empty_data = parse_data([])
    get_top_student(empty_data)
```

```python
# Custom Exceptions
class InvalidTicketClassError(Exception):
    """Raised when an invalid ticket class is provided"""
    def __init__(self, ticket_class):
        self.ticket_class = ticket_class
        super().__init__(f"Invalid ticket class: '{ticket_class}'. Must be 'Economy', 'Business', or 'First Class'")

class InvalidLuggageWeightError(Exception):
    """Raised when luggage weight is invalid"""
    def __init__(self, weight):
        self.weight = weight
        super().__init__(f"Invalid luggage weight: {weight} kg. Weight must be positive.")

def calculate_fare(ticket_class, luggage_weight):
    """
    Calculate total fare based on ticket class and luggage weight

    Args:
        ticket_class (str): 'Economy', 'Business', or 'First Class'
        luggage_weight (float): Weight of luggage in kg

    Returns:
        int: Total fare in Rs.

    Raises:
        InvalidTicketClassError: If invalid ticket class provided
        InvalidLuggageWeightError: If luggage weight is negative
    """
    # Define base fares
    base_fares = {
        'Economy': 15000,
        'Business': 30000,
        'First Class': 50000
    }

    # Validate ticket class
    if ticket_class not in base_fares:
        raise InvalidTicketClassError(ticket_class)

    # Validate luggage weight
    if luggage_weight < 0:
        raise InvalidLuggageWeightError(luggage_weight)

    # Calculate base fare
    base_fare = base_fares[ticket_class]

    # Calculate luggage charges (above 20kg)
    free_luggage_allowance = 20
    if luggage_weight > free_luggage_allowance:
        extra_weight = luggage_weight - free_luggage_allowance
        luggage_charge = extra_weight * 500
    else:
        luggage_charge = 0

    total_fare = base_fare + luggage_charge

    return total_fare

def generate_invoice(passenger_name, ticket_class, luggage_weight):
    """
    Generate an invoice for airline ticket with proper exception handling

    Args:
        passenger_name (str): Name of the passenger
        ticket_class (str): Ticket class
        luggage_weight (float): Luggage weight in kg

    Returns:
        int: Total fare or None if error occurred
    """
    print(f"\nüé´ PROCESSING TICKET FOR: {passenger_name}")
    print("=" * 50)

    try:
        # Calculate fare
        total_fare = calculate_fare(ticket_class, luggage_weight)

        # Additional calculations for invoice details
        base_fares = {'Economy': 15000, 'Business': 30000, 'First Class': 50000}
        base_fare = base_fares[ticket_class]
        luggage_charge = max(0, (luggage_weight - 20) * 500)

    except InvalidTicketClassError as e:
        print(f"‚ùå Error: {e}")
        print("üí° Please choose from: Economy, Business, or First Class")
        return None

    except InvalidLuggageWeightError as e:
        print(f"‚ùå Error: {e}")
        print("üí° Luggage weight cannot be negative")
        return None

    except Exception as e:
        print(f"üí• Unexpected error: {e}")
        return None

    else:
        # This runs only if no exception was raised
        print("‚úÖ Ticket processed successfully!")
        print("\nüìÑ INVOICE DETAILS:")
        print(f"   Passenger: {passenger_name}")
        print(f"   Class: {ticket_class}")
        print(f"   Luggage Weight: {luggage_weight} kg")
        print(f"   Base Fare: Rs.{base_fare:,}")
        if luggage_charge > 0:
            print(f"   Extra Luggage Charge: Rs.{luggage_charge:,}")
        print(f"   {'-' * 30}")
        print(f"   TOTAL FARE: Rs.{total_fare:,}")

        return total_fare

    finally:
        # This always runs, regardless of exception
        print("\n" + "=" * 50)
        print("üìã Thank you for using Airline Ticketing System!")
        print("=" * 50)

def bulk_ticket_processing(bookings):
    """
    Process multiple tickets in bulk
    """
    print("üöÄ BULK TICKET PROCESSING")
    print("=" * 60)

    successful_bookings = 0
    total_revenue = 0

    for i, (name, ticket_class, weight) in enumerate(bookings, 1):
        print(f"\nüì¶ Processing Booking {i}:")
        fare = generate_invoice(name, ticket_class, weight)

        if fare is not None:
            successful_bookings += 1
            total_revenue += fare

    print(f"\nüìä BULK PROCESSING SUMMARY:")
    print(f"   Successful bookings: {successful_bookings}/{len(bookings)}")
    print(f"   Total revenue: Rs.{total_revenue:,}")

# Demonstration
if __name__ == "__main__":
    print("‚úàÔ∏è  AIRLINE TICKET FARE CALCULATOR")
    print("=" * 60)

    # Test cases for individual bookings
    test_cases = [
        # (passenger_name, ticket_class, luggage_weight)
        ("Ali Khan", "Economy", 15),
        ("Sara Ahmed", "Business", 25),
        ("Hassan Raza", "First Class", 30),
        ("Zara Malik", "Premium Economy", 20),  # Invalid class
        ("Bilal Shah", "Business", -5),         # Invalid weight
        ("Ayesha Noor", "First Class", 18)
    ]

    # Process individual test cases
    for passenger_name, ticket_class, luggage_weight in test_cases:
        fare = generate_invoice(passenger_name, ticket_class, luggage_weight)
        if fare:
            print(f"üí∞ Fare for {passenger_name}: Rs.{fare:,}")

    # Bulk processing demonstration
    print("\n" + "=" * 70)
    bulk_bookings = [
        ("John Doe", "Economy", 22),
        ("Jane Smith", "Business", 18),
        ("Bob Wilson", "First Class", 35),
        ("Alice Brown", "Economy", 19),
        ("Charlie Davis", "Business", 28)
    ]

    bulk_ticket_processing(bulk_bookings)

    # Interactive example
    print("\n" + "=" * 70)
    print("üíª INTERACTIVE EXAMPLE")
    print("=" * 70)

    # Simulate user input
    demo_data = [
        ("Demo Passenger", "Business", 24)
    ]

    for name, t_class, weight in demo_data:
        fare = generate_invoice(name, t_class, weight)
        if fare:
            print(f"üéØ Final fare calculated: Rs.{fare:,}")
```

```python
# Custom Exception for invalid ratings
class InvalidRatingError(Exception):
    """Raised when employee rating is outside valid range (1-5)"""
    def __init__(self, employee_name, rating):
        self.employee_name = employee_name
        self.rating = rating
        super().__init__(f"Invalid rating {rating} for {employee_name}. Rating must be between 1-5.")

def calculate_bonus(employee_dict):
    """
    Calculate bonuses for employees based on salary and performance rating

    Args:
        employee_dict (dict): Dictionary with employee names as keys and
                             tuples (salary, rating) as values

    Returns:
        dict: Dictionary mapping employee names to bonus amounts
    """

    def bonus_message(name, bonus):
        """
        Inner function to print bonus message
        """
        print(f"üéâ {name} will receive a bonus of {bonus:,.0f} PKR.")

    print("üí∞ EMPLOYEE BONUS DISTRIBUTION SYSTEM")
    print("=" * 50)

    bonus_dict = {}
    successful_calculations = 0
    failed_calculations = 0

    for employee, (salary, rating) in employee_dict.items():
        try:
            # Validate salary
            if salary <= 0:
                raise ValueError(f"Invalid salary {salary:,} PKR for {employee}. Salary must be positive.")

            # Validate rating
            if not (1 <= rating <= 5):
                raise InvalidRatingError(employee, rating)

            # Check if rating is integer
            if not isinstance(rating, int):
                raise ValueError(f"Rating must be integer for {employee}. Got {type(rating).__name__}.")

            # Calculate bonus: bonus = salary √ó (rating / 10)
            bonus = salary * (rating / 10)
            bonus_dict[employee] = bonus
            successful_calculations += 1

            # Use inner function to print message
            bonus_message(employee, bonus)

        except InvalidRatingError as e:
            failed_calculations += 1
            print(f"‚ùå {e}")

        except ValueError as e:
            failed_calculations += 1
            print(f"‚ùå {e}")

        except Exception as e:
            failed_calculations += 1
            print(f"üí• Unexpected error processing {employee}: {e}")

    print("=" * 50)
    print(f"üìä PROCESSING SUMMARY:")
    print(f"   Successful calculations: {successful_calculations}")
    print(f"   Failed calculations: {failed_calculations}")
    print(f"   Total employees processed: {len(employee_dict)}")

    return bonus_dict

def generate_bonus_report(bonus_dict, employee_dict):
    """
    Generate a comprehensive bonus report
    """
    if not bonus_dict:
        print("üì≠ No valid bonus data to report")
        return

    print("\nüìà BONUS DISTRIBUTION REPORT")
    print("=" * 60)

    # Calculate statistics
    total_bonus = sum(bonus_dict.values())
    avg_bonus = total_bonus / len(bonus_dict)
    max_bonus_emp = max(bonus_dict.items(), key=lambda x: x[1])
    min_bonus_emp = min(bonus_dict.items(), key=lambda x: x[1])

    print(f"Total Bonus Distributed: {total_bonus:,.0f} PKR")
    print(f"Average Bonus: {avg_bonus:,.0f} PKR")
    print(f"Highest Bonus: {max_bonus_emp[0]} - {max_bonus_emp[1]:,.0f} PKR")
    print(f"Lowest Bonus: {min_bonus_emp[0]} - {min_bonus_emp[1]:,.0f} PKR")

    print("\n" + "üìã DETAILED BREAKDOWN:")
    print("-" * 40)
    for employee, bonus in sorted(bonus_dict.items(), key=lambda x: x[1], reverse=True):
        salary, rating = employee_dict[employee]
        print(f"  {employee}:")
        print(f"    Salary: {salary:,.0f} PKR")
        print(f"    Rating: {rating}/5")
        print(f"    Bonus: {bonus:,.0f} PKR (={salary:,.0f} √ó {rating}/10)")
        print()

# Using dictionary comprehension for alternative implementation
def calculate_bonus_comprehension(employee_dict):
    """
    Alternative implementation using dictionary comprehension with error handling
    """
    def calculate_single_bonus(employee, salary, rating):
        """Helper function to calculate bonus for one employee"""
        try:
            if not (1 <= rating <= 5):
                raise InvalidRatingError(employee, rating)
            if salary <= 0:
                raise ValueError(f"Invalid salary for {employee}")
            return salary * (rating / 10)
        except (InvalidRatingError, ValueError):
            return None

    # Dictionary comprehension with error handling
    bonus_dict = {
        employee: calculate_single_bonus(employee, salary, rating)
        for employee, (salary, rating) in employee_dict.items()
        if calculate_single_bonus(employee, salary, rating) is not None
    }

    return bonus_dict

# Demonstration
if __name__ == "__main__":
    # Test data with various scenarios
    employee_data = {
        'Ali': (50000, 4),
        'Sara': (60000, 5),
        'Hassan': (55000, 7),      # Invalid rating
        'Zara': (48000, 3),
        'Bilal': (52000, 2),
        'Ayesha': (58000, 6),      # Invalid rating
        'Raza': (45000, 1),
        'Fatima': (-50000, 4),     # Invalid salary
        'John': (60000, 4.5),      # Non-integer rating
        'Amir': (47000, 0),        # Invalid rating
        'Nadia': (51000, 5)
    }

    print("üè¢ EMPLOYEE BONUS CALCULATION SYSTEM")
    print("=" * 70)

    # Method 1: Using detailed error handling
    print("\n1Ô∏è‚É£  MAIN CALCULATION METHOD:")
    bonus_results = calculate_bonus(employee_data)

    # Generate comprehensive report
    generate_bonus_report(bonus_results, employee_data)

    # Method 2: Using dictionary comprehension
    print("\n2Ô∏è‚É£  ALTERNATIVE METHOD (Dictionary Comprehension):")
    print("=" * 50)
    bonus_comp = calculate_bonus_comprehension(employee_data)

    if bonus_comp:
        print("‚úÖ Valid bonuses calculated using comprehension:")
        for employee, bonus in bonus_comp.items():
            print(f"   {employee}: {bonus:,.0f} PKR")
    else:
        print("üì≠ No valid bonuses could be calculated")

    # Display final results
    print("\n" + "=" * 70)
    print("üéØ FINAL BONUS DISTRIBUTION:")
    if bonus_results:
        for employee, bonus in sorted(bonus_results.items()):
            salary, rating = employee_data[employee]
            print(f"   {employee}: Salary {salary:,} √ó Rating {rating}/5 = Bonus {bonus:,.0f} PKR")
    else:
        print("   No valid bonus data available")

    # Calculate total successful payout
    if bonus_results:
        total_payout = sum(bonus_results.values())
        print(f"\nüí∞ TOTAL BONUS PAYOUT: {total_payout:,.0f} PKR")
        print(f"üë• EMPLOYEES RECEIVING BONUS: {len(bonus_results)}")
```

```python
# Custom Exception for student not found (optional enhancement)
class StudentNotFoundError(Exception):
    """Raised when student is not found in the records"""
    def __init__(self, student_name):
        self.student_name = student_name
        super().__init__(f"Student '{student_name}' not found in records.")

def get_student_grade(student_name, grades_dict):
    """
    Get a student's grade from the grades dictionary

    Args:
        student_name (str): Name of the student to search for
        grades_dict (dict): Dictionary with student names as keys and grades as values

    Returns:
        str: Student's grade if found

    Raises:
        KeyError: If student name doesn't exist in dictionary
    """
    print(f"\nüîç Searching for student: '{student_name}'")

    try:
        # Attempt to access the student's grade
        grade = grades_dict[student_name]
        return grade

    except KeyError:
        # Handle the case where student is not found
        print(f"‚ùå KeyError: Student '{student_name}' not found in records.")
        raise  # Re-raise the exception to let caller handle it

    finally:
        # This block always executes, regardless of success or failure
        print("‚úÖ Search completed.")

def get_student_grade_enhanced(student_name, grades_dict):
    """
    Enhanced version with custom exception and additional validation
    """
    print(f"\nüîç ENHANCED SEARCH for: '{student_name}'")

    try:
        # Additional validation
        if not isinstance(grades_dict, dict):
            raise TypeError("grades_dict must be a dictionary")

        if not student_name or not isinstance(student_name, str):
            raise ValueError("Student name must be a non-empty string")

        # Attempt to access the student's grade
        if student_name not in grades_dict:
            raise StudentNotFoundError(student_name)

        grade = grades_dict[student_name]
        return grade

    except StudentNotFoundError as e:
        print(f"‚ùå {e}")
        raise

    except TypeError as e:
        print(f"‚ùå TypeError: {e}")
        raise

    except ValueError as e:
        print(f"‚ùå ValueError: {e}")
        raise

    except Exception as e:
        print(f"üí• Unexpected error: {e}")
        raise

    finally:
        print("‚úÖ Search completed.")

def demonstrate_grade_lookup():
    """
    Demonstrate the grade lookup functionality with various test cases
    """
    # Sample grades dictionary
    grades_data = {
        'Ali': 'A',
        'Sara': 'B+',
        'Hassan': 'A-',
        'Zara': 'C+',
        'Bilal': 'B',
        'Ayesha': 'A+'
    }

    print("üéì STUDENT GRADE LOOKUP SYSTEM")
    print("=" * 50)
    print("Available students:", list(grades_data.keys()))
    print("=" * 50)

    # Test cases: both existing and non-existing students
    test_students = ['Ali', 'Sara', 'John', 'Zara', 'Michael', 'Ayesha']

    for student in test_students:
        try:
            grade = get_student_grade(student, grades_data)
            print(f"üéØ Found: {student} ‚Üí Grade: {grade}")

        except KeyError:
            print(f"üìù Action: Student '{student}' will be added to missing records.")

        print("-" * 40)

    # Demonstrate enhanced version
    print("\n" + "=" * 60)
    print("ENHANCED VERSION DEMONSTRATION")
    print("=" * 60)

    enhanced_test_cases = [
        ('Hassan', grades_data),      # Valid case
        ('Unknown', grades_data),     # Non-existent student
        ('Ali', "not_a_dict"),        # Invalid dictionary
        ('', grades_data),            # Empty student name
    ]

    for student, data in enhanced_test_cases:
        try:
            grade = get_student_grade_enhanced(student, data)
            print(f"üéØ Found: {student} ‚Üí Grade: {grade}")

        except (StudentNotFoundError, TypeError, ValueError) as e:
            print(f"üìù Handling error: {type(e).__name__}")

        except Exception as e:
            print(f"üí• Critical error: {e}")

        print("-" * 40)

def bulk_student_lookup(student_names, grades_dict):
    """
    Demonstrate bulk lookup with exception handling
    """
    print("\n" + "=" * 50)
    print("BULK STUDENT LOOKUP")
    print("=" * 50)

    found_count = 0
    not_found_count = 0

    for student in student_names:
        try:
            grade = get_student_grade(student, grades_dict)
            found_count += 1
            print(f"   ‚úÖ {student}: {grade}")

        except KeyError:
            not_found_count += 1
            print(f"   ‚ùå {student}: NOT FOUND")

    print(f"\nüìä BULK LOOKUP SUMMARY:")
    print(f"   Found: {found_count} students")
    print(f"   Not Found: {not_found_count} students")
    print(f"   Total: {len(student_names)} students")

# Explanation of how exception handling ensures program stability
def explain_exception_handling_benefits():
    """
    Explain the importance of exception handling in dictionary lookups
    """
    print("\n" + "=" * 70)
    print("üìö HOW EXCEPTION HANDLING ENSURES PROGRAM STABILITY")
    print("=" * 70)

    benefits = {
        "1. Prevents Program Crashes":
            "Without try-except, a KeyError would crash the entire program. With exception handling, the program continues running.",

        "2. Graceful Error Recovery":
            "The program can recover from errors and continue processing other students instead of stopping completely.",

        "3. Clean Resource Management":
            "The 'finally' block ensures that cleanup code always runs, regardless of success or failure.",

        "4. Meaningful Error Messages":
            "Custom error messages help users understand what went wrong and how to fix it.",

        "5. Maintains Data Integrity":
            "By handling exceptions properly, we prevent partial updates or corrupted data states.",

        "6. Enables Bulk Processing":
            "We can process large datasets where some entries might be invalid without stopping the entire operation.",

        "7. Better Debugging":
            "Structured exception handling makes it easier to identify and fix issues in the code.",

        "8. User-Friendly Experience":
            "Users see helpful messages instead of scary Python tracebacks."
    }

    for benefit, explanation in benefits.items():
        print(f"\n{benefit}:")
        print(f"   {explanation}")

# Main demonstration
if __name__ == "__main__":
    # Sample data
    grades_db = {
        'Ali': 'A',
        'Sara': 'B+',
        'Hassan': 'A-',
        'Zara': 'C+',
        'Bilal': 'B',
        'Ayesha': 'A+',
        'Fatima': 'B-',
        'Imran': 'C'
    }

    # Run demonstrations
    demonstrate_grade_lookup()

    # Bulk lookup demonstration
    bulk_students = ['Ali', 'Sara', 'John', 'Zara', 'Mike', 'Ayesha', 'Fatima', 'David']
    bulk_student_lookup(bulk_students, grades_db)

    # Educational explanation
    explain_exception_handling_benefits()

print("\nNote: Exception handling transforms fragile code that crashes on errors into robust, production-ready code that gracefully handles unexpected situations!")
```

```python
def merge_sets(set1, set2):
    """
    Merge two sets and return their union with type validation

    Args:
        set1: First set to merge
        set2: Second set to merge

    Returns:
        set: Union of set1 and set2, or empty set if error occurs

    Raises:
        TypeError: If either input is not a set
    """
    print(f"\nüîÑ Attempting to merge: {set1} and {set2}")
    print(f"   set1 type: {type(set1).__name__}, set2 type: {type(set2).__name__}")

    try:
        # Type validation
        if not isinstance(set1, set) or not isinstance(set2, set):
            raise TypeError("Both inputs must be sets.")

        # Perform the union operation
        result = set1 | set2  # Using union operator
        # Alternative: result = set1.union(set2)

        print(f"‚úÖ Successfully merged sets!")
        return result

    except TypeError as e:
        print(f"‚ùå TypeError: {e}")
        return set()  # Return empty set as error recovery

    except Exception as e:
        print(f"üí• Unexpected error: {e}")
        return set()

def merge_sets_enhanced(set1, set2):
    """
    Enhanced version with detailed error messages and automatic conversion
    """
    print(f"\nüîß ENHANCED: Merging {set1} and {set2}")

    try:
        # Detailed type validation with specific error messages
        if not isinstance(set1, set):
            raise TypeError(f"First argument must be a set, got {type(set1).__name__}")
        if not isinstance(set2, set):
            raise TypeError(f"Second argument must be a set, got {type(set2).__name__}")

        result = set1 | set2
        print(f"‚úÖ Enhanced merge successful!")
        return result

    except TypeError as e:
        print(f"‚ùå Enhanced TypeError: {e}")

        # Attempt automatic conversion for some types
        try:
            print("üîÑ Attempting automatic conversion...")
            converted_set1 = set(set1) if not isinstance(set1, set) else set1
            converted_set2 = set(set2) if not isinstance(set2, set) else set2
            result = converted_set1 | converted_set2
            print(f"‚úÖ Auto-conversion successful: {result}")
            return result
        except:
            print("‚ùå Auto-conversion failed")
            return set()

    except Exception as e:
        print(f"üí• Unexpected error in enhanced version: {e}")
        return set()

def demonstrate_merge_operations():
    """
    Demonstrate various merge scenarios with different data types
    """
    print("üéØ SET MERGING DEMONSTRATION")
    print("=" * 60)

    # Test cases with various data types
    test_cases = [
        # Valid cases
        ({1, 2, 3}, {3, 4, 5}),
        (set(), {1, 2, 3}),  # Empty set
        ({'a', 'b'}, {'b', 'c'}),

        # Invalid cases - wrong types
        ([1, 2, 3], {4, 5, 6}),  # List instead of set
        ({1, 2}, (3, 4, 5)),     # Tuple instead of set
        ("hello", {1, 2, 3}),    # String instead of set
        (123, {1, 2, 3}),        # Integer instead of set
        ({1, 2}, [3, 4]),        # Both wrong types
        ({"a": 1}, {1, 2}),      # Dict instead of set
    ]

    for i, (arg1, arg2) in enumerate(test_cases, 1):
        print(f"\n{'='*50}")
        print(f"TEST CASE {i}:")
        result = merge_sets(arg1, arg2)
        print(f"üì¶ Result: {result}")

    # Demonstrate enhanced version
    print("\n" + "=" * 70)
    print("ENHANCED VERSION DEMONSTRATION")
    print("=" * 70)

    enhanced_cases = [
        ({1, 2, 3}, [4, 5, 6]),  # Auto-convertible
        ((1, 2, 3), {4, 5, 6}),  # Auto-convertible
        ("abc", {1, 2, 3}),      # Partially convertible
    ]

    for i, (arg1, arg2) in enumerate(enhanced_cases, 1):
        print(f"\n{'='*50}")
        print(f"ENHANCED TEST {i}:")
        result = merge_sets_enhanced(arg1, arg2)
        print(f"üì¶ Final result: {result}")

def explain_type_checking_importance():
    """
    Explain why type checking is important in Python's dynamic typing system
    """
    print("\n" + "=" * 80)
    print("üìö IMPORTANCE OF TYPE CHECKING IN PYTHON'S DYNAMIC TYPING SYSTEM")
    print("=" * 80)

    benefits = {
        "1. Prevents Runtime Errors":
            "Dynamic typing allows any variable to hold any type, but operations are type-specific. "
            "Without validation, you might try to call set methods on a list, causing AttributeError at runtime.",

        "2. Improves Code Reliability":
            "Type checking acts as a guard clause, ensuring functions receive expected inputs. "
            "This prevents subtle bugs that might not surface immediately.",

        "3. Enhances Debugging":
            "Early type validation provides clear error messages pointing to the exact problem, "
            "rather than cryptic errors deep in the function logic.",

        "4. Better Documentation":
            "Type checking serves as executable documentation, making function requirements explicit "
            "and helping other developers understand how to use your code.",

        "5. Maintains Data Integrity":
            "Ensuring correct types prevents data corruption and maintains consistency "
            "throughout your application's data flow.",

        "6. Facilitates Refactoring":
            "With type validation, you can confidently refactor code knowing that type mismatches "
            "will be caught early rather than causing runtime failures.",

        "7. Improves User Experience":
            "Clear type error messages help users of your functions understand what went wrong "
            "and how to fix their input data.",

        "8. Enables Polymorphism Safely":
            "You can still support multiple types when appropriate, but do so explicitly "
            "rather than accidentally accepting wrong types."
    }

    for benefit, explanation in benefits.items():
        print(f"\n{benefit}:")
        print(f"   {explanation}")

    # Practical examples
    print("\n" + "=" * 50)
    print("üîç PRACTICAL EXAMPLES")
    print("=" * 50)

    print("\nWithout type checking:")
    print("  merge_sets([1,2,3], {4,5,6}) would fail with:")
    print("  AttributeError: 'list' object has no attribute 'union'")

    print("\nWith type checking:")
    print("  merge_sets([1,2,3], {4,5,6}) gives:")
    print("  TypeError: Both inputs must be sets.")

    print("\nDynamic Typing Challenge:")
    print("  Python allows: x = {1,2,3}; x = [1,2,3]; x = 'hello'")
    print("  But: x.union() works only when x is a set!")

# Performance comparison
def demonstrate_performance_impact():
    """
    Show the performance and reliability impact of type checking
    """
    print("\n" + "=" * 60)
    print("‚ö° PERFORMANCE AND RELIABILITY IMPACT")
    print("=" * 60)

    def unsafe_merge(set1, set2):
        """No type checking - might fail at runtime"""
        return set1 | set2

    def safe_merge(set1, set2):
        """With type checking - fails early with clear message"""
        if not isinstance(set1, set) or not isinstance(set2, set):
            raise TypeError("Both inputs must be sets")
        return set1 | set2

    # Test both approaches
    test_data = ({1, 2, 3}, {3, 4, 5})
    invalid_data = ([1, 2, 3], {3, 4, 5})

    print("Testing with valid data:")
    try:
        result1 = unsafe_merge(*test_data)
        result2 = safe_merge(*test_data)
        print(f"‚úÖ Both work: {result1}, {result2}")
    except Exception as e:
        print(f"Error: {e}")

    print("\nTesting with invalid data:")
    try:
        unsafe_merge(*invalid_data)
    except Exception as e:
        print(f"‚ùå Unsafe version fails: {type(e).__name__}: {e}")

    try:
        safe_merge(*invalid_data)
    except Exception as e:
        print(f"‚ùå Safe version fails: {type(e).__name__}: {e}")

# Main execution
if __name__ == "__main__":
    # Run all demonstrations
    demonstrate_merge_operations()
    demonstrate_performance_impact()
    explain_type_checking_importance()

print("\nNote: Type checking in Python's dynamic system is crucial for building reliable, maintainable software. It transforms runtime errors into predictable, manageable validation failures with clear messaging!")
```

```python
def safe_tuple_access(data_tuple, index):
    """
    Safely retrieve an element from a tuple at the specified index

    Args:
        data_tuple: Tuple to access
        index: Index position to retrieve

    Returns:
        Element at the specified index, or None if error occurs

    Raises:
        TypeError: If data_tuple is not a tuple
        IndexError: If index is out of range
    """
    print(f"\nüîç Attempting to access index {index} in tuple: {data_tuple}")

    try:
        # Validate input type
        if not isinstance(data_tuple, tuple):
            raise TypeError(f"Expected tuple, got {type(data_tuple).__name__}")

        # Validate index type
        if not isinstance(index, int):
            raise TypeError(f"Index must be an integer, got {type(index).__name__}")

        # Attempt to access the element
        element = data_tuple[index]
        print(f"‚úÖ Successfully accessed: data_tuple[{index}] = {element}")
        return element

    except TypeError as e:
        print(f"‚ùå TypeError: {e}")
        return None

    except IndexError as e:
        print(f"‚ùå IndexError: Invalid index access. Tuple has {len(data_tuple)} elements, index {index} is out of range")
        return None

    except Exception as e:
        print(f"üí• Unexpected error: {e}")
        return None

def safe_tuple_access_enhanced(data_tuple, index, default_value="DEFAULT"):
    """
    Enhanced version with default value and detailed error handling
    """
    print(f"\nüîß ENHANCED: Accessing index {index} in {data_tuple}")

    try:
        # Comprehensive type validation
        if not isinstance(data_tuple, tuple):
            raise TypeError(f"First argument must be a tuple, got {type(data_tuple).__name__}")

        if not isinstance(index, int):
            raise TypeError(f"Index must be an integer, got {type(index).__name__}")

        # Check for negative indices
        if index < 0:
            if abs(index) > len(data_tuple):
                raise IndexError(f"Negative index {index} out of range for tuple of length {len(data_tuple)}")
            print(f"‚ö†Ô∏è  Using negative index: {index} (equivalent to position {len(data_tuple) + index})")

        element = data_tuple[index]
        print(f"‚úÖ Enhanced access successful: {element}")
        return element

    except TypeError as e:
        print(f"‚ùå Enhanced TypeError: {e}")
        print(f"   Returning default value: {default_value}")
        return default_value

    except IndexError as e:
        print(f"‚ùå Enhanced IndexError: {e}")
        print(f"   Tuple length: {len(data_tuple)}, Valid indices: 0 to {len(data_tuple)-1}")
        print(f"   Returning default value: {default_value}")
        return default_value

    except Exception as e:
        print(f"üí• Enhanced unexpected error: {e}")
        return default_value

def demonstrate_tuple_access():
    """
    Demonstrate various tuple access scenarios with different data types and indices
    """
    print("üéØ SAFE TUPLE ACCESS DEMONSTRATION")
    print("=" * 60)

    # Test cases with various scenarios
    test_cases = [
        # Valid cases
        (('apple', 'banana', 'cherry'), 1),
        ((10, 20, 30, 40, 50), 0),
        (('single',), 0),
        ((1, 2, 3), -1),  # Negative index
        ((1, 2, 3), -2),  # Negative index

        # Invalid cases - wrong types
        (['list', 'instead', 'of', 'tuple'], 1),  # List instead of tuple
        (('a', 'b', 'c'), "string_index"),        # String index
        (('x', 'y'), 2.5),                        # Float index
        (123, 0),                                 # Integer instead of tuple
        ("string", 1),                            # String instead of tuple

        # Invalid cases - out of range
        (('a', 'b', 'c'), 5),                     # Index too high
        (('a', 'b', 'c'), -5),                    # Negative index too low
        ((), 0),                                  # Empty tuple
        (('single',), 1),                         # Index beyond single element
    ]

    for i, (data, idx) in enumerate(test_cases, 1):
        print(f"\n{'='*50}")
        print(f"TEST CASE {i}:")
        result = safe_tuple_access(data, idx)
        print(f"üì¶ Result: {result}")

    # Demonstrate enhanced version
    print("\n" + "=" * 70)
    print("ENHANCED VERSION DEMONSTRATION")
    print("=" * 70)

    enhanced_cases = [
        (('apple', 'banana', 'cherry'), 2),
        (('a', 'b'), 5),           # Out of range
        ([1, 2, 3], 1),            # Wrong type
        (('x', 'y'), "invalid"),   # Invalid index type
    ]

    for i, (data, idx) in enumerate(enhanced_cases, 1):
        print(f"\n{'='*50}")
        print(f"ENHANCED TEST {i}:")
        result = safe_tuple_access_enhanced(data, idx, "NOT_FOUND")
        print(f"üì¶ Final result: {result}")

def demonstrate_program_robustness():
    """
    Show how safe tuple access prevents program crashes
    """
    print("\n" + "=" * 70)
    print("üõ°Ô∏è  DEMONSTRATING PROGRAM ROBUSTNESS")
    print("=" * 70)

    sample_data = ('Alice', 'Bob', 'Charlie', 'Diana')

    print("Sample data:", sample_data)
    print("Length:", len(sample_data))
    print("Valid indices: 0 to", len(sample_data) - 1)

    print("\n" + "-" * 50)
    print("WITHOUT EXCEPTION HANDLING (CRASHES):")
    print("-" * 50)

    # This would crash the program
    try:
        # Simulating unsafe access
        print("Attempting unsafe access...")
        risky_result = sample_data[10]  # This would cause IndexError
        print("This line never executes if IndexError occurs")
    except IndexError as e:
        print(f"‚ùå Program would crash here: {e}")
        print("üí• Entire program stops if not handled!")

    print("\n" + "-" * 50)
    print("WITH EXCEPTION HANDLING (CONTINUES):")
    print("-" * 50)

    # Safe access continues program execution
    test_indices = [0, 2, 5, -1, -6]  # Mix of valid and invalid indices

    for idx in test_indices:
        result = safe_tuple_access(sample_data, idx)
        if result is not None:
            print(f"   ‚úÖ Using result: {result}")
        else:
            print(f"   ‚ö†Ô∏è  Handling missing data gracefully")

    print("\n‚úÖ Program continues executing normally after errors!")

def explain_robustness_benefits():
    """
    Explain how exception handling promotes program robustness
    """
    print("\n" + "=" * 80)
    print("üìö HOW EXCEPTION HANDLING PROMOTES PROGRAM ROBUSTNESS")
    print("=" * 80)

    benefits = {
        "1. Prevents Program Crashes":
            "Without exception handling, an IndexError or TypeError would terminate the entire program. "
            "With proper handling, the program continues running and can recover gracefully.",

        "2. Graceful Error Recovery":
            "Instead of crashing, the function returns a sensible default value (None or custom default) "
            "allowing the calling code to decide how to handle the situation.",

        "3. User-Friendly Experience":
            "Users see clear error messages explaining what went wrong, rather than scary Python tracebacks "
            "that reveal implementation details.",

        "4. Maintains Data Integrity":
            "By catching errors early and handling them appropriately, we prevent partial or corrupted "
            "data states that could cause cascading failures.",

        "5. Enables Bulk Processing":
            "When processing large datasets, individual errors don't stop the entire operation. "
            "The system can log errors and continue with remaining data.",

        "6. Better Debugging":
            "Structured error handling provides context about where and why errors occurred, "
            "making debugging much easier than searching through crash logs.",

        "7. Production Reliability":
            "Robust error handling is essential for production systems where programs must run "
            "continuously without manual intervention.",

        "8. Defensive Programming":
            "Anticipating and handling potential errors makes code more resilient to unexpected "
            "inputs and edge cases."
    }

    for benefit, explanation in benefits.items():
        print(f"\nüõ°Ô∏è  {benefit}:")
        print(f"   {explanation}")

    # Code comparison
    print("\n" + "=" * 60)
    print("üîç CODE COMPARISON: ROBUST vs FRAGILE")
    print("=" * 60)

    print("\nFRAGILE CODE (crashes on error):")
    fragile_code = """
def fragile_access(data, index):
    return data[index]  # Crashes if index out of range

# These would crash the program:
fragile_access(('a', 'b'), 5)      # IndexError
fragile_access([1,2,3], 1)         # Works but wrong type
fragile_access(('x',), "invalid")  # TypeError
"""
    print(fragile_code)

    print("\nROBUST CODE (handles errors gracefully):")
    robust_code = """
def robust_access(data, index):
    try:
        if not isinstance(data, tuple):
            raise TypeError("Expected tuple")
        return data[index]
    except (TypeError, IndexError) as e:
        print(f"Error: {e}")
        return None

# These are handled gracefully:
robust_access(('a', 'b'), 5)      # Returns None with error message
robust_access([1,2,3], 1)         # Returns None with error message
robust_access(('x',), "invalid")  # Returns None with error message
"""
    print(robust_code)

# Real-world application example
def process_student_records():
    """
    Demonstrate real-world application of safe tuple access
    """
    print("\n" + "=" * 70)
    print("üéì REAL-WORLD APPLICATION: STUDENT RECORDS PROCESSING")
    print("=" * 70)

    # Simulating student records from a database
    student_records = [
        ('Alice', 'A', 95, 'CS'),
        ('Bob', 'B', 85, 'Math'),
        ('Charlie', 'A', 92, 'Physics'),
        # Sometimes data might be incomplete or malformed
        ('Diana',),  # Incomplete record
        ('Eve', 'C', 78),
        (123, 'B', 82, 'Bio'),  # Wrong data type
    ]

    print("Processing student records...")
    print("Each record format: (name, grade, score, department)")
    print("-" * 50)

    for i, record in enumerate(student_records, 1):
        print(f"\nRecord {i}: {record}")

        # Safely access each field
        name = safe_tuple_access_enhanced(record, 0, "Unknown")
        grade = safe_tuple_access_enhanced(record, 1, "Ungraded")
        score = safe_tuple_access_enhanced(record, 2, 0)
        department = safe_tuple_access_enhanced(record, 3, "Undeclared")

        # Process the data
        if name != "Unknown" and grade != "Ungraded":
            print(f"   ‚úÖ Processed: {name} - Grade: {grade}, Score: {score}, Dept: {department}")
        else:
            print(f"   ‚ö†Ô∏è  Skipped incomplete record")

    print("\n‚úÖ All records processed without crashes!")

# Main execution
if __name__ == "__main__":
    # Run all demonstrations
    demonstrate_tuple_access()
    demonstrate_program_robustness()
    process_student_records()
    explain_robustness_benefits()

print("\n 4. Production System Benefits:")
print("\n24/7 Operation: Programs don't crash due to unexpected inputs")
print("\nAutomated Recovery: Systems can continue after temporary issues")
print("\nBetter Monitoring: Structured error handling enables better logging and alerting")
print("\nMaintainability: Clear error paths make debugging and maintenance easier")
print("\nNote: Exception handling transforms fragile code that crashes on unexpected inputs into robust, production-ready software that gracefully handles errors and continues operation. This is essential for building reliable systems that users can depend on!")
```
