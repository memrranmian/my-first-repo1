```python
# -----------------------------------------------------------------------------
# House Price Prediction using Simple Linear Regression
# Goal: To predict a house price (Y) based on a single feature, the size (X).
# -----------------------------------------------------------------------------

# Import necessary libraries
# 1. pandas: For handling and displaying data in a table format (DataFrame).
# 2. numpy: For efficient numerical operations and array handling.
# 3. matplotlib.pyplot: For creating visualizations (the graph).
# 4. LinearRegression from sklearn: The core linear regression model.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# --- 1. Define the Dataset ---
# We use a small, simple (synthetic) dataset for this beginner example.
# X: Feature (Input) - Size of the house in square feet (sqft).
# Y: Target (Output) - Price of the house in thousands of dollars ($k).
house_sizes_sqft = np.array([1200, 1450, 1600, 1800, 2000, 2300, 2500, 2800])
house_prices_k = np.array([250, 300, 320, 370, 410, 480, 520, 550])

# Reshape the input array (X) to be 2D, as required by scikit-learn
# The original shape is (n_samples,), we need (n_samples, n_features)
X = house_sizes_sqft.reshape(-1, 1)
Y = house_prices_k

# --- 2. Create and Display a DataFrame (The Table) ---
print("="*50)
print("üè° Initial House Data Table (DataFrame)")
print("="*50)

# Create a Pandas DataFrame for clear, tabular display
data = pd.DataFrame({
    'Size (sqft)': house_sizes_sqft,
    'Price (k$)': house_prices_k
})
print(data)
print("\n")


# --- 3. Initialize and Train the Model ---
print("="*50)
print("ü§ñ Model Training (Simple Linear Regression)")
print("="*50)

# Initialize the Linear Regression model.
# Linear Regression models the relationship between a dependent variable (Y, Price)
# and one or more independent variables (X, Size) by fitting a linear equation.
model = LinearRegression()

# Train the model using the .fit() method.
# The model finds the best-fit straight line (Y = m*X + b) that minimizes the
# distance between the line and the actual data points.
model.fit(X, Y)

# Retrieve the model parameters (m and b)
# Coefficient (m or slope): How much Y changes for a one-unit change in X.
slope = model.coef_[0]
# Intercept (b): The predicted value of Y when X is 0.
intercept = model.intercept_

print(f"Model trained successfully.")
print(f"  Coefficient (Slope, m): {slope:.2f}")
print(f"  Intercept (b): {intercept:.2f}")
print(f"  The formula found is: Price (k$) = {slope:.2f} * Size (sqft) + {intercept:.2f}")


# --- 4. Prediction Example ---
# Use the trained model to predict the price for a new, unseen house size.
new_house_size = 2100 # A house of 2100 sqft
# The input for prediction must also be a 2D array.
X_predict = np.array([[new_house_size]])

# Use the .predict() method to get the predicted price (Y_predict)
Y_predict = model.predict(X_predict)
predicted_price = Y_predict[0]

print("\n" + "="*50)
print("üéØ Prediction for New Data")
print("="*50)
print(f"Input House Size: {new_house_size} sqft")
print(f"Predicted Price: ${predicted_price:.2f} k") # e.g., $430.50 k
print("\n")


# --- 5. Visualization (Plotting the Results) ---
print("üìä Generating Visualization...")

# Generate the regression line values for plotting
# We use the trained model to predict Y values for all X inputs.
Y_line = model.predict(X)

plt.figure(figsize=(10, 6))

# Plot the original data points (Scatter Plot)
# These are the actual recorded prices.
plt.scatter(house_sizes_sqft, house_prices_k, color='blue', label='Actual House Prices', s=100)

# Plot the Regression Line
# This is the "best-fit" line found by the Linear Regression algorithm.
plt.plot(house_sizes_sqft, Y_line, color='red', label='Linear Regression Line')

# Plot the Predicted Datum ('X')
# This is the price prediction for the new_house_size (2100 sqft).
plt.scatter(new_house_size, predicted_price, color='green', marker='x', s=200, label=f'Predicted Price ({predicted_price:.2f} k$)', linewidths=3)

# Add Labels and Title
plt.title('House Price Prediction using Linear Regression')
plt.xlabel('House Size (Square Feet)')
plt.ylabel('Price (Thousands of $)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

print("\nScript execution finished. Check the generated graph window.")
```
